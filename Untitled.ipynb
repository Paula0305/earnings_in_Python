```python
import requests
r = requests.get("https://justjoin.it/api/offers")
dataset = r.json()
highest_b2b = 0
lowest_b2b = 0
highest_per = 0
lowest_per = 0
counter_b2b = 0
counter_permanent = 0
value = []


for offer in dataset:
    for empt in offer['employment_types']:

        if empt['type'] == 'b2b':
            if empt['salary'] != None:
                value.append(empt['salary']['currency'])                
                highest_b2b += empt['salary']['to']
                lowest_b2b += empt['salary']['from']
                counter_b2b += 1
                
        elif empt['type'] == 'permanent':
            if empt['salary'] != None:
                value.append(empt['salary']['currency'])
                highest_per += empt['salary']['to']
                lowest_per += empt['salary']['from']
                counter_permanent += 1

                          
print('Średnia najniższa kwota b2b: ', round(lowest_b2b / counter_b2b, 2), "zł") 
print('Średnia najwyższa kwota b2b: ', round(highest_b2b / counter_b2b, 2), "zł") 
print('Średnia najniższa kwota permanent: ', round(lowest_per / counter_permanent,2), "zł") 
print('Średnia najwyższa kwota permanent: ', round(highest_per / counter_permanent, 2), "zł") 
print(set(value))
```
